### Task 2: Enhancing Scatter Plots


#### 1. Scatter Plot with Trend Line for Age vs. Job Satisfaction



Add a regression line to the scatter plot of Age vs. JobSatPoints_6 to highlight trends in the data.


```python
# 任務 2：增強散佈圖 
# 1. 年齡與工作滿意度趨勢線散佈圖 在年齡與 JobSatPoints_6 散佈圖中加入迴歸線，以突顯資料中的趨勢。
#
```

```python
## Write your code here

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert Age to numeric
age_mapping = {
    'Under 18 years old': 16,  # Approximate value
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70  # Approximate value
}

df['Age_Numeric'] = df['Age'].map(age_mapping)

# Remove rows with NaN (if any mapping failed)
df = df.dropna(subset=['Age_Numeric', 'JobSatPoints_6'])

# Create scatter plot with trend line
plt.figure(figsize=(10, 6))
sns.regplot(x='Age_Numeric', y='JobSatPoints_6', data=df, scatter_kws={'alpha': 0.5})
plt.title('Scatter Plot of Age vs. Job Satisfaction with Trend Line')
plt.xlabel('Age (Numeric)')
plt.ylabel('Job Satisfaction (JobSatPoints_6)')
plt.show()
```

#### 2. Scatter Plot for Age vs. Work Experience


Visualize the relationship between Age (`Age`) and Work Experience (`YearsCodePro`) using a scatter plot.


```python
# 2. 年齡與工作經驗的散佈圖 
# 使用散佈圖視覺化年齡（Age）和工作經驗（YearsCodePro）之間的關係。
#
```

```python
## Write your code here
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Convert Age to numeric
age_mapping = {
    'Under 18 years old': 16,  # Approximate value
    '18-24 years old': 21,
    '25-34 years old': 30,
    '35-44 years old': 40,
    '45-54 years old': 50,
    '55-64 years old': 60,
    '65 years or older': 70  # Approximate value
}

df['Age_Numeric'] = df['Age'].map(age_mapping)

# Clean and convert 'YearsCodePro' to numeric
def clean_years_code(years):
    try:
        return float(years)
    except ValueError:
        return None

df['YearsCodePro_Numeric'] = df['YearsCodePro'].apply(clean_years_code)

# Remove rows with NaN
df = df.dropna(subset=['Age_Numeric', 'YearsCodePro_Numeric'])

# Create scatter plot
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age_Numeric', y='YearsCodePro_Numeric', data=df)
plt.title('Scatter Plot of Age vs. Work Experience')
plt.xlabel('Age (Numeric)')
plt.ylabel('Years of Professional Coding Experience (Numeric)')
plt.show()
```

### Task 3: Combining Scatter Plots with Additional Features


#### 1. Bubble Plot of Compensation vs. Job Satisfaction with Age as Bubble Size



Create a bubble plot to explore the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSatPoints_6`), with bubble size representing age.


```python
# 任務 3：將散佈圖與附加特徵結合 
# 1. 薪資與工作滿意度的氣泡圖，以年齡為氣泡大小
#
# 建立氣泡圖來探討年度薪資（ConvertedCompYearly）和工作滿意度（JobSatPoints_6）之間的關係，氣泡大小代表年齡。
#
```

#### 2. Scatter Plot for Popular Programming Languages by Job Satisfaction


Visualize the popularity of programming languages (`LanguageHaveWorkedWith`) against job satisfaction using a scatter plot. Use points to represent satisfaction levels for each language.


```python
# 2. 熱門程式語言工作滿意度散佈圖
# 使用散點圖直觀地展示程式語言（LanguageHaveWorkedWith）的流行度與工作滿意度之間的關係。用點數來表示每種語言的滿意度。
#
```

```python
## Write your code here    


# import pandas as pd
# import matplotlib.pyplot as plt
# import seaborn as sns
#
# file_path = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_path)

# Clean LanguageHaveWorkedWith and JobSatPoints_6
df = df.dropna(subset=['LanguageHaveWorkedWith', 'JobSatPoints_6'])

# Split the languages and create a new DataFrame
languages = df['LanguageHaveWorkedWith'].str.split(';', expand=True).stack().reset_index(level=1, drop=True).rename('Language')
df_languages = pd.merge(languages, df['JobSatPoints_6'], left_index=True, right_index=True)

# Get the top 10 languages by popularity
top_languages = df_languages['Language'].value_counts().nlargest(10).index.tolist()

# Filter for top 10 languages
df_top_languages = df_languages[df_languages['Language'].isin(top_languages)]

# Calculate the mean job satisfaction for each language
language_satisfaction = df_top_languages.groupby('Language')['JobSatPoints_6'].mean().reset_index()

# Create scatter plot
plt.figure(figsize=(14, 8))
sns.scatterplot(x='Language', y='JobSatPoints_6', data=language_satisfaction)
plt.title('Scatter Plot of Popular Programming Languages by Job Satisfaction')
plt.xlabel('Programming Language')
plt.ylabel('Mean Job Satisfaction (JobSatPoints_6)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
